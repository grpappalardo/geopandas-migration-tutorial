```
# code and package recommendations from Data Camp tutorial
import numpy as np
import pandas as pd
import geopandas
from shapely.geometry import Point
import csv
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
% matplotlib inline
```

```
# source https://datahub.io/core/geo-countries
# view top of data
countries = geopandas.read_file("countries.geojson")
countries.head()
```

```
# #htmlcolorcodes.com for hex codes
#plot countries
countries[countries['ADMIN'].isin(['Antarctica']) == False].plot(figsize=(50,40), color='#6493DF')
```

```
# code adapted from Data Camp tutorial
# limit to just Russia
countries[countries['ADMIN'].isin(['Russia']) == True].plot(figsize=(50,40), color='#18A351')
```

```
# limit to just Kazakhstan
countries[countries['ADMIN'].isin(['Kazakhstan']) == True].plot(figsize=(10, 10), color='#2786C4')
```

```
from google.colab import files
uploaded = files.upload()
# data from Federal Reserve Bank of St. Louis

# code from https://www.pythonforengineers.com/introduction-to-pandas/
# view Russian migration csv
russiadata = pd.read_csv('SMPOPNETMRUS.csv')
russiadata.head()
```

```
# shows the pandas data type
type(russiadata)
```

```
russiadata.set_index("DATE", inplace = True) # replaces the indexing with the date information as x axis
russiadata.head()
```

```
# plot graph showing trends (up )
russiadata.plot(kind='bar',color='#18A351')
plt.title('Russian Migration 1962-2017')
plt.ylabel('In and Out Flow')
plt.show()
```
